## Merging Phase

- By merging the airline dataframe into the flight dataframe as the flight dataframe has lots of attributes so this dataframe will be treated as the main one.
- AIRLINE is the attribute thats added to the flight dataframe from merging the airline dataframe into the flight dataframe.
- Merging will be by using the merge on arguments type of using the left_on and right_on features for merging tables on diffrent name of columns but same values.

```python
merged_dataframes = pd.merge(flights_data,airlines_data, left_on='AIRLINE' , right_on= 'IATA_CODE')
```

```python
merged_dataframes.head()
```

```python
# Droping the attribute as its values are already exiting in the table under another name of attrbute 
merged_dataframes = merged_dataframes.drop(['IATA_CODE'], axis=1)
```

```python
merged_dataframes.head()
```

- Here is the merging between the merged dataframe and the airport data frame to add these attributes:
    - AIRPORT
    - CITY
    - STATE
    - COUNTRY
    - LATITUDE
    - LONGTITUDE

```python
merged_dataframes = pd.merge(merged_dataframes,airports_data, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE')
```

```python
merged_dataframes.head()
```

- Here is dealing with the repeated attributs with same repeated columns for better clean results.

```python
merged_dataframes = merged_dataframes.drop(['IATA_CODE_y'], axis=1)
```

```python
merged_dataframes.head()
```

```python
merged_dataframes = merged_dataframes.drop(['STATE_x','COUNTRY_x','LATITUDE_x','LONGITUDE_x',], axis=1)
```

```python
merged_dataframes.head()
```

```python
merged_dataframes.head()
```

```python
merged_dataframes = merged_dataframes.drop(['IATA_CODE_x'], axis=1)
```

```python
merged_dataframes.head()
```

- Renaming the columns to its original names for clean looking.

```python
merged_dataframes.rename(columns = {'AIRPORT_y':'AIRPORT','CITY_y':'CITY','STATE_y':'STATE','COUNTRY_y':'COUNTRY','LATITUDE_y':'LATITUDE','LONGITUDE_y':'LONGITUDE'}, inplace = True)
```
