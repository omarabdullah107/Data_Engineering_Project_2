### Using the box plot

- Box plot will be used for handling the outliers detecting them and then removing them.

```python
# Function for creating the box plot
def plot_boxplot(df,ft):
    df.boxplot(column=[ft])
    plt.grid(False)
    plt.show()

def outliers(df,ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    ls = df.index[ (df[ft] < lower_bound) | (df[ft] > upper_bound)]
    
    return ls

# Function for removing the outliers

def remove(df,ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```

### What year and month had the most flights?

### What airport had the most flights in 2015-01-01?

```python
# Filter the data to include only flights on a specific date
df = merged_dataframes_new[merged_dataframes_new['DATE'] == '2015-01-01']

# Group the data by 'airport' and count the number of flights for each airport
df_by_airport = df.groupby('AIRPORT').FLIGHT_NUMBER.count()

# Sort the data by the number of flights in descending order
df_by_airport = df_by_airport.sort_values(ascending=False)

# Select the top 20 airports
df_top_15 = df_by_airport[:20]

# Plot the data as a bar chart
df_top_15.plot(kind='bar', x='AIRPORT', y='FLIGHT_NUMBER', title='Number of Flights by Airport')

# Show the plot
plt.show()
```

- well appearing from this graph that Hartsfield-Jackson Atlanta International Airport is the airport that got the most flights within the the first 20 airports had flights in the date of 2015-01-01, we chose 20 airports because there are lots of airports which make the x-line so messy with data and not readable.
