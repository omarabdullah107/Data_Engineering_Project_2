### Using the box plot

- Box plot will be used for handling the outliers detecting them and then removing them.

```python
# Function for creating the box plot
def plot_boxplot(df,ft):
    df.boxplot(column=[ft])
    plt.grid(False)
    plt.show()

def outliers(df,ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    ls = df.index[ (df[ft] < lower_bound) | (df[ft] > upper_bound)]
    
    return ls

# Function for removing the outliers

def remove(df,ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```
